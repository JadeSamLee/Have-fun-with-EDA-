# Data Distribution and Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plot the distribution of the target variable
plt.figure(figsize=(8, 6))
sns.countplot(data=application_data, x='AMT_INCOME_TOTAL')
plt.title('Distribution of Target Variable')
plt.show()
```

```python
# Distribution of income
plt.figure(figsize=(8, 6))
sns.histplot(application_data['AMT_INCOME_TOTAL'], bins=50, kde=True)
plt.title('Distribution of Annual Income')
plt.show()
```

```python
# Boxplot for income by target
plt.figure(figsize=(10, 6))
sns.boxplot(data=application_data, x='DAYS_EMPLOYED', y='AMT_INCOME_TOTAL')
plt.title('Income Distribution by Target')
plt.show()
```

```python
# Countplot for categorical variables
plt.figure(figsize=(12, 10))
sns.countplot(data=application_data, x='CODE_GENDER')
plt.title('Distribution of Gender')
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.countplot(data=application_data, x='NAME_EDUCATION_TYPE')
plt.title('Distribution of Education Level')
plt.show()
```
